```python
import sys
import os
from pathlib import Path

# 获取当前 notebook 所在的目录（Jupyter 中）
try:
    # 尝试获取 __file__（在 .py 文件中有效）
    current_dir = os.path.dirname(os.path.abspath(__file__))
except NameError:
    # 在 Jupyter notebook 中，使用当前工作目录
    current_dir = os.getcwd()
    
# 将当前目录添加到系统路径
sys.path.append(current_dir)

from delayed_model import DelayedStateTransitionModel
from utils.helpers import format_state_sequence, save_simulation_result
from base_model import EventType
```

```python
def demo_delayed_model():
    """演示延迟状态转移模型 - 中间状态不可观测"""
    print("=" * 80)
    print("延迟状态转移模型演示（中间状态不可观测）")
    print("=" * 80)
    
    # 创建模型实例
    model = DelayedStateTransitionModel(n=2, m=4, k=2)
    
    # 查看模型信息
    info = model.get_model_info()
    print(f"\n📊 模型信息:")
    print(f"   名称: {info['model_name']}")
    print(f"   参数: n={info['model_params']['n']}, m={info['model_params']['m']}, k={info['model_params']['k']}")
    print(f"   初始状态: {info['initial_state']} (不健康)")
    print(f"   可用干预: {info['available_interventions']}")
    print(f"\n⚠️  注意: 中间状态是不可观测的潜在状态，不会出现在输出中")
    
    # 定义干预序列
    interventions = [
        (0, 0, 1),  # step 1: 施加干预
        (0, 0, 1),  # step 2: 继续干预（达到n=2，进入部分转移，但观测状态仍为不健康）
        (0, 0, 0),  # step 3: 停止干预（仍在中间状态持续期内）
        (0, 0, 0),  # step 4: 仍然没有干预（中间状态持续期结束，仍为不健康）
        (0, 0, 0),  # step 5: 无干预
        (0, 0, 1),  # step 6: 重新干预
        (0, 0, 1),  # step 7: 继续干预
        (0, 0, 1),  # step 8: 继续干预
        (0, 0, 1),  # step 9: 继续干预（达到m=4，完全转移到健康）
    ]
    
    print(f"\n📋 干预序列（共 {len(interventions)} 步）:")
    for t, inter in enumerate(interventions, 1):
        print(f"   t={t}: {inter}", end="")
        if inter == (0, 0, 1):
            print(" 🔵 施加干预")
        else:
            print(" ⚪ 无干预")
    
    # 模拟
    result = model.simulate(interventions, return_details=True)
    
    # 格式化输出 - 详细版
    print(f"\n📈 模拟结果（可观测状态序列）:")
    print("-" * 90)
    print(f"{'步数':<6} {'干预向量':<25} {'可观测状态':<25} {'事件类型':<30}")
    print("-" * 90)
    
    for i in range(len(result["states"])):
        if i == 0:
            inter_str = "初始状态"
        else:
            inter_str = str(result["interventions"][i-1])
        
        state_str = str(result["states"][i])
        
        # 添加状态说明
        if result["states"][i] == model.UNHEALTHY:
            state_desc = "不健康"
        elif result["states"][i] == model.HEALTHY:
            state_desc = "健康"
        else:
            state_desc = "未知"
        
        display_state = f"{state_str} ({state_desc})"
        
        if i > 0 and "events" in result:
            event = result["events"][i-1]
            # 添加事件的中文说明
            if event == EventType.COMPLETE_TRANSITION:
                event_str = f"{event.value} (完全转移 → 健康)"
            elif event == EventType.PARTIAL_TRANSITION:
                event_str = f"{event.value} (部分转移 → 进入不可观测中间状态)"
            elif event == EventType.NO_EFFECTIVE_TRANSITION:
                if model._partial_transition_counter > 0:
                    event_str = f"{event.value} (仍在中间状态持续期，剩余{model._partial_transition_counter}步)"
                else:
                    event_str = f"{event.value} (无有效干预 → 不健康)"
            else:
                event_str = event.value
        else:
            event_str = "-"
        
        print(f"{i:<6} {inter_str:<25} {display_state:<25} {event_str:<30}")
    
    print("-" * 90)
    
    # 添加状态图例
    print(f"\n📖 可观测状态说明:")
    print(f"   (0, 0, 1) → 不健康状态")
    print(f"   (0, 0, 0) → 健康状态")
    
    # 添加不可观测状态说明
    print(f"\n🔒 不可观测的潜在状态:")
    print(f"   部分转移后，模型内部会进入一个持续 {model.model_params['k']} 步的中间状态")
    print(f"   在此期间，即使停止干预，观测到的状态仍然是不健康")
    print(f"   只有完全转移才能直接到达健康状态")
    
    # 保存结果
    save_simulation_result(result, "delayed_model_result.json")
    print(f"\n💾 结果已保存到 delayed_model_result.json")
    
    return model, result
```

```python
def demo_comparison():
    """对比演示：展示不可观测中间状态的影响"""
    print("\n" + "=" * 80)
    print("对比演示：部分转移的不可观测性")
    print("=" * 80)
    
    model = DelayedStateTransitionModel(n=2, m=4, k=3)
    
    print(f"\n模型参数: n={model.model_params['n']}, m={model.model_params['m']}, k={model.model_params['k']}")
    print(f"说明: 连续 {model.model_params['n']} 步干预后进入中间状态（不可观测），持续 {model.model_params['k']} 步")
    
    # 场景：部分转移后停止干预
    interventions = [(0, 0, 1)] * 2 + [(0, 0, 0)] * 5
    
    print(f"\n干预序列: {interventions}")
    print("\n从外部观察者的角度看：")
    
    result = model.simulate(interventions, return_details=True)
    
    print(f"\n{'步数':<6} {'干预':<12} {'观察到的状态':<20} {'实际内部状态':<25}")
    print("-" * 70)
    
    # 手动跟踪内部状态用于演示
    internal_state = "不健康"
    partial_counter = 0
    
    for i in range(len(result["states"])):
        if i == 0:
            inter = "初始"
            observed = result["states"][i]
            observed_str = "不健康" if observed == model.UNHEALTHY else "健康"
            internal_str = "不健康"
        else:
            inter = "干预" if interventions[i-1] == model.INTERVENTION else "无干预"
            observed = result["states"][i]
            observed_str = "不健康" if observed == model.UNHEALTHY else "健康"
            
            # 模拟内部状态逻辑
            if i <= 2:  # 前2步连续干预
                if i == 2:  # 达到n=2，进入中间状态
                    partial_counter = model.model_params["k"]
                    internal_str = f"中间状态（剩余{partial_counter}步）"
                else:
                    internal_str = "不健康（积累干预）"
            else:  # 停止干预后
                if partial_counter > 0:
                    partial_counter -= 1
                    if partial_counter > 0:
                        internal_str = f"中间状态（剩余{partial_counter}步）"
                    else:
                        internal_str = "不健康（中间状态结束）"
                else:
                    internal_str = "不健康"
        
        print(f"{i:<6} {inter:<12} {observed_str:<20} {internal_str:<25}")
    
    print("\n💡 结论: 从外部观察者的角度看，状态始终是'不健康'")
    print("   只有通过后续行为（如继续干预到完全转移）才能观察到状态变化")
```

```python
if __name__ == "__main__" or True:
    # 运行主演示
    model, result = demo_delayed_model()
    
    # 运行对比演示
    demo_comparison()
    
    print("\n" + "=" * 80)
    print("✅ 所有演示完成！")
    print("=" * 80)
```
